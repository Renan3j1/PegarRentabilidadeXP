```python
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options 
from selenium.webdriver.support import expected_conditions as EC
import time
import urllib
from selenium.webdriver.chrome.service import Service
import random
import numpy as np
```

Confirmar tabela

```python
ID = "id"
XPATH = "xpath"

contatos_data = pd.read_excel("Pronto_Envio_Abril_2025.xlsx")
contatos_data = contatos_data.fillna("")

contatos_data['rentabilidade'] = contatos_data['rentabilidade'].str.replace("_x000D_\n", "")
contatos_data['lucro'] = contatos_data['lucro'].str.replace("_x000D_\n", "")
contatos_data['rentabilidade'] = contatos_data['rentabilidade'].str.replace("\r\n", "")
contatos_data['lucro'] = contatos_data['lucro'].str.replace("\r\n", "")
contatos_data['cdi'] = contatos_data['cdi'].str.replace("_x000D_\n", "")
contatos_data['cdi'] = contatos_data['cdi'].str.replace("\r\n", "")
contatos_data 
```

Iniciar o navegador

```python
service = Service(ChromeDriverManager().install())
options = Options()
options.add_experimental_option("detach", True)

navegador = webdriver.Chrome(service=service, options=options)

navegador.get("https://web.whatsapp.com/")

while len(navegador.find_elements(By.ID, "side")) < 1:
    time.sleep(5)
```

Rodar as mensagens
